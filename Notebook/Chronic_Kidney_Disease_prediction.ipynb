```python
import pandas as pd
import numpy as np
```

```python
dt= pd.read_csv('../DATA/kidney_disease.csv')
```

```python
# dropping id column
dt.drop('id', axis = 1, inplace = True)
```

```python
dt.columns = ['age', 'blood_pressure', 'specific_gravity', 'albumin', 'sugar', 'red_blood_cells', 'pus_cell',
              'pus_cell_clumps', 'bacteria', 'blood_glucose_random', 'blood_urea', 'serum_creatinine', 'sodium',
              'potassium', 'haemoglobin', 'packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count',
              'hypertension', 'diabetes_mellitus', 'coronary_artery_disease', 'appetite', 'peda_edema',
              'aanemia', 'class']
```

```python
dt.head()
```

```python
dt.info()
```

```python
# converting necessary columns to numerical type

dt['packed_cell_volume'] = pd.to_numeric(dt['packed_cell_volume'], errors='coerce')
dt['white_blood_cell_count'] = pd.to_numeric(dt['white_blood_cell_count'], errors='coerce')
dt['red_blood_cell_count'] = pd.to_numeric(dt['red_blood_cell_count'], errors='coerce')
```

```python
# Extracting categorical and numerical columns

cat_cols = [col for col in dt.columns if dt[col].dtype == 'object']
num_cols = [col for col in dt.columns if dt[col].dtype != 'object']
```

```python
# looking at unique values in categorical columns

for col in cat_cols:
    print(f"{col} has {dt[col].unique()} values\n")
```

```python
# replace incorrect values

dt['diabetes_mellitus'].replace(to_replace = {'\tno':'no','\tyes':'yes',' yes':'yes'},inplace=True)

dt['coronary_artery_disease'] = dt['coronary_artery_disease'].replace(to_replace = '\tno', value='no')

dt['class'] = dt['class'].replace(to_replace = {'ckd\t': 'ckd', 'notckd': 'not ckd'})
```

```python
dt['class'] = dt['class'].map({'ckd': 0, 'not ckd': 1})
dt['class'] = pd.to_numeric(dt['class'], errors='coerce')
```

```python
cols = ['diabetes_mellitus', 'coronary_artery_disease', 'class']

for col in cols:
    print(f"{col} has {dt[col].unique()} values\n")
```

```python
# filling null values, we will use two methods, random sampling for higher null values and 
# mean/mode sampling for lower null values

def random_value_imputation(feature):
    random_sample = dt[feature].dropna().sample(dt[feature].isna().sum())
    random_sample.index = dt[dt[feature].isnull()].index
    dt.loc[dt[feature].isnull(), feature] = random_sample
    
def impute_mode(feature):
    mode = dt[feature].mode()[0]
    dt[feature] = dt[feature].fillna(mode)
```

```python
# filling num_cols null values using random sampling method

for col in num_cols:
    random_value_imputation(col)
```

```python
transformed_data = onehotencoder.fit_transform(dt[cat_cols])
```

```python
cat_columns= ['red_blood_cells_abnormal','red_blood_cells_normal','pus_cell_abnormal','pus_cell_normal','pus_cell_clumps_notpresent',
             'pus_cell_clumps_present','bacteria_notpresent','bacteria_present','hypertension_no','hypertension_yes','diabetes_mellitus_no',
             'diabetes_mellitus_yes','coronary_artery_disease_no','coronary_artery_disease_yes','appetite_good','appetite_poor','peda_edema_no',
             'peda_edema_yes','aanemia_no','aanemia_yes']
```

```python
transformed_data = pd.DataFrame(transformed_data,index=dt.index,columns = cat_columns)
```

```python
pd.DataFrame(transformed_data).head(10)
```

```python
X = pd.concat([dt[num_cols],transformed_data],axis = 1)
```

```python
X.columns
```

```python
y = dt[['class']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=3)
```

```python
#Applying decision tree model using hyper parameter optimization using grid serach CV
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Number of trees in random forest
n_estimators = [400,500,600]
# Number of features to consider at every split
max_features = [6,8,10]
# Maximum number of levels in tree
max_depth = [4,6,8,10]
# Minimum number of samples required to split a node
min_samples_split = [10,20]
# Minimum number of samples required at each leaf node
min_samples_leaf = [10,20]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
param_dist = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
# Instantiating Decision Tree classifier 
rf = RandomForestClassifier() 
  
# Instantiating GridSearchCV object 
rf_cv = GridSearchCV(rf, param_dist, cv = 3)
```

```python
#fitting X and y
rf_cv.fit(X_train, y_train)
```

```python
# Print the tuned parameters and score 
print("Tuned Random Forest Parameters: {}".format(rf_cv.best_params_)) 
print("Best score is {}".format(rf_cv.best_score_))
```

```python
new_rf = RandomForestClassifier(bootstrap = False, max_depth = 4, max_features = 6, min_samples_leaf = 10, min_samples_split = 10, n_estimators = 500) 
```

```python
new_rf.fit(X_train, y_train)
```

```python
predictions_rf = new_rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(classification_report(y_test,predictions_rf))
```
