```python
import pandas as pd
import numpy as np
```

```python
dt = pd.read_csv('../DATA/parkinsons.csv')
```

```python
dt.head()
```

```python
dt.isnull().sum()
```

```python
dt['status'].value_counts()
```

```python
X = dt[dt.columns.difference(['name', 'status'])]
```

```python
X.columns
```

```python
y = dt[['status']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.40, random_state=3)
```

```python
#Applying decision tree model using hyper parameter optimization using grid serach CV
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Number of trees in random forest
n_estimators = [400,500,600]
# Number of features to consider at every split
max_features = [6,8,10]
# Maximum number of levels in tree
max_depth = [4,6,8,10]
# Minimum number of samples required to split a node
min_samples_split = [6,10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [6,10]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
param_dist = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
# Instantiating Decision Tree classifier 
rf = RandomForestClassifier() 
  
# Instantiating GridSearchCV object 
rf_cv = GridSearchCV(rf, param_dist, cv = 3)
```

```python
#fitting X and y
rf_cv.fit(X_train, y_train)
```

```python
# Print the tuned parameters and score 
print("Tuned Random Forest Parameters: {}".format(rf_cv.best_params_)) 
print("Best score is {}".format(rf_cv.best_score_))
```

```python
predictions_rf = rf_cv.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(classification_report(y_test,predictions_rf))
```

```python
# loading library
import pickle

# create an iterator object with write permission - model.pkl
with open('../models/parkinsons/parkinsons_model.pkl', 'wb') as files:
    pickle.dump(rf_cv, files)
```
