```python
import pandas as pd
import numpy as np
```

```python
dt = pd.read_csv('../DATA/HepatitisCdata.csv')
```

```python
dt.head()
```

```python
dt.info()
```

```python
dt.columns
```

```python
dt['Category'] = dt['Category'].map({'0=Blood Donor':0,'0s=suspect Blood Donor': 0,'1=Hepatitis': 1,
                                    '2=Fibrosis': 1,'3=Cirrhosis':1})
```

```python
dt['Sex'] = dt['Sex'].map({'m':0,'f':1})
```

```python
dt.isnull().sum()
```

```python
def random_value_imputation(feature):
    random_sample = dt[feature].dropna().sample(dt[feature].isna().sum())
    random_sample.index = dt[dt[feature].isnull()].index
    dt.loc[dt[feature].isnull(), feature] = random_sample
```

```python
random_value_imputation('ALB')
random_value_imputation('ALP')
random_value_imputation('ALT')
random_value_imputation('CHOL')
random_value_imputation('PROT')
```

```python
dt.columns
```

```python
X = dt[['Age', 'Sex', 'ALB', 'ALP', 'ALT', 'AST', 'BIL', 'CHE',
       'CHOL', 'CREA', 'GGT', 'PROT']]
```

```python
y = dt[['Category']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=3)
```

```python
#Applying decision tree model using hyper parameter optimization using grid serach CV
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Number of trees in random forest
n_estimators = [400,500,600]
# Number of features to consider at every split
max_features = [6,8,10]
# Maximum number of levels in tree
max_depth = [4,6,8,10]
# Minimum number of samples required to split a node
min_samples_split = [10,20]
# Minimum number of samples required at each leaf node
min_samples_leaf = [10,20]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
param_dist = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
# Instantiating Decision Tree classifier 
rf = RandomForestClassifier() 
  
# Instantiating GridSearchCV object 
rf_cv = GridSearchCV(rf, param_dist, cv = 3)
```

```python
#fitting X and y
rf_cv.fit(X_train, y_train)
```

```python
# Print the tuned parameters and score 
print("Tuned Random Forest Parameters: {}".format(rf_cv.best_params_)) 
print("Best score is {}".format(rf_cv.best_score_))
```

```python
new_rf = RandomForestClassifier(bootstrap = False, max_depth = 4, max_features = 6, min_samples_leaf = 10, min_samples_split = 10, n_estimators = 500) 
```

```python
new_rf.fit(X_train, y_train)
```

```python
import shap

explainer = shap.TreeExplainer(new_rf, X_train)
```

```python
predictions_rf = rf_cv.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(classification_report(y_test,predictions_rf))
```

```python
# loading library
import pickle

# create an iterator object with write permission - model.pkl
with open('../models/hepatitisC/hepatitisC_model.pkl', 'wb') as files:
    pickle.dump(new_rf, files)  

# create an iterator object with write permission - model.pkl
with open('../models/hepatitisC/hepatitisC_explainer.pkl', 'wb') as files:
    pickle.dump(explainer, files)
```
